```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
```

```python
df = pd.read_csv(r"C:\Users\giftt\Downloads\sneakers_streetwear_sales_data.csv")
df.head()
```

```python
print(df.info())
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
df_clean = df.drop_duplicates()
print(f"Removed {len(df) - len(df_clean)} duplicate rows")
```

```python
display(df.describe())
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Category', palette='pastel')

plt.title('SALES COUNT BY CATEGORY')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
revenue_by_gender = df.groupby('Gender')['Amount ($)'].sum().sort_values()

plt.figure(figsize=(6, 4))
sns.barplot(x=revenue_by_gender.values, y=revenue_by_gender.index)
plt.title('Revenue by Gender')
plt.xlabel('Total Revenue ($)')
plt.ylabel('Gender')
plt.tight_layout()
plt.show()
```

```python
quantity_by_gender = df.groupby('Gender')['Quantity'].sum().sort_values()

plt.figure(figsize=(6, 4))
sns.barplot(x=quantity_by_gender.values, y=quantity_by_gender.index)
plt.title('Quantity Sold by Gender')
plt.xlabel('Total Quantity')
plt.ylabel('Gender')
plt.tight_layout()
plt.show()
```

```python
avg_price_by_type = df.groupby('Product Type')['Unit Price ($)'].mean().sort_values()

plt.figure(figsize=(8, 5))
sns.barplot(x=avg_price_by_type.values, y=avg_price_by_type.index)
plt.title('Average Unit Price by Product Type')
plt.xlabel('Average Price ($)')
plt.ylabel('Product Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='Unit Price ($)', y='Quantity', hue='Category')
plt.title('Unit Price vs Quantity Sold')
plt.xlabel('Unit Price ($)')
plt.ylabel('Quantity Sold')
plt.tight_layout()
plt.show()
```

```python
top_products = df.groupby('Product Name')['Amount ($)'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 Products by Revenue')
plt.xlabel('Total Revenue ($)')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, y='Payment Mode', order=df['Payment Mode'].value_counts().index)
plt.title('Payment Mode Distribution')
plt.xlabel('Number of Transactions')
plt.ylabel('Payment Mode')
plt.tight_layout()
plt.show()
```

```python
revenue_by_category = df.groupby('Category')['Amount ($)'].sum().sort_values()
plt.figure(figsize=(8, 5))
sns.barplot(x=revenue_by_category.values, y=revenue_by_category.index)
plt.title('Revenue by Category')
plt.xlabel('Total Revenue ($)')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load dataset
df = pd.read_csv(r"C:\Users\giftt\Downloads\sneakers_streetwear_sales_data.csv")

# Parse 'Date' column to datetime and set as index
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])  # Remove rows where 'Date' couldn't be parsed
df.set_index('Date', inplace=True)

# Set seaborn theme
sns.set_theme(style='whitegrid', palette='Set2')

# Resample monthly (month-end) and plot
monthly_sales = df['Amount ($)'].resample('ME').sum()

plt.figure(figsize=(10, 5))
monthly_sales.plot(marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
